# 4. 스타일

```python
#폰트 디폴트 설정하면 좋은 것들
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'AppleGothic' #맥 설정
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시, 마이너스 글자가 꺠지는거 방지
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y, linewidth = 10) # 선 두께 설정
```

## Marker

```python
plt.plot(x,y, marker = 'o') #데이터가 있는 부분에 표시
```

```python
plt.plot(x,y, marker = 'o', linestyle = 'None') #데이터가 있는 부분에 표시
```

```python
plt.plot(x,y, marker = 'v', markersize = 10) #데이터가 있는 부분에 표시
```


marker 종류 등이 있음

```python
plt.plot(x,y, marker = 'v', markersize = 10, markeredgecolor = 'red')
```

```python
plt.plot(x,y, marker = 'v', markersize = 10, markeredgecolor = 'red', markerfacecolor = 'red')
```

## 선 스타일

```python
plt.plot(x,y,linestyle = '--')
```

```python
# 다양한 선 종류 존재 구글 검색
```

# 색깔

```python
plt.plot(x,y, color ='pink')
```

```python
#구글에서 검색!
```

# 포맷

```python
plt.plot(x,y, 'ro--') #color, marker, linestyle순으로 작성
```

```python
plt.plot(x,y,'go') #라인 스타일을 지정하지 않음으로써 아무 선도 없다
```

# 축약어

```python
plt.plot(x,y, marker = 'o', mfc = 'r',ms = 10, mec = 'blue', ls = ':')
```

# 투명도

```python
plt.plot(x,y,marker = 'o', mfc = 'red', ms = 10, alpha = 0.3) #투명도는 알파로 설정 (0~1)의 값으로 설정!
```

## 그래프 크기

```python
plt.figure(figsize = (10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize = (10,5), dpi = 400) # dots per inch, 해상도 확대
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor = 'yellow')
plt.plot(x,y)
```
