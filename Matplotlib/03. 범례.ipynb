# 3. 범례(legend)

```python
#폰트 디폴트 설정하면 좋은 것들
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'AppleGothic' #맥 설정
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시, 마이너스 글자가 꺠지는거 방지
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y, label = '추천 데이터')
plt.legend()
```

```python
plt.plot(x,y, label = '추천 데이터')
plt.legend(loc = 'upper right') # 범례 위치 설정
```

```python
plt.plot(x,y, label = '추천 데이터')
plt.legend(loc =(0.5,0.5)) # x,y축 기준 (0~1 사이)
```
