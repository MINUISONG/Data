# 02. 축

```python
#폰트 디폴트 설정하면 좋은 것들
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'AppleGothic' #맥 설정
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시, 마이너스 글자가 꺠지는거 방지
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.title('꺾은선 그래프', fontdict = {'family': 'Gungseo', 'size': 20}) #개별 폰트 설정
```

```python
#x, y축에 대한 정보 출력
plt.plot(x,y)
plt.xlabel('x축', color = 'red')
plt.ylabel('y축')
```

```python
#x, y축에 대한 정보 출력, 레이블의 위치를 설정
plt.plot(x,y)
plt.xlabel('x축', color = 'red', loc = 'right') #right, bottom, center
plt.ylabel('y축', loc = 'top') #top, center, bottom
```

```python
#그래프의 원하는 틱들을 출력
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.ylim(1, 12)  # Y축 범위를 3에서 12로 설정
plt.plot(x,y)
plt.show()
```
