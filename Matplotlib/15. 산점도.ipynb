```python
#폰트 디폴트 설정하면 좋은 것들
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'AppleGothic' #맥 설정
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시, 마이너스 글자가 꺠지는거 방지
```

```python
import pandas as pd
df= pd.read_excel('../Pandas/socre.xlsx')
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s = sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500
plt.scatter(df['영어'], df['수학'], s = sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500
plt.scatter(df['영어'], df['수학'], s = sizes, c = df['학년'], cmap = 'viridis', alpha = 0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500
plt.scatter(df['영어'], df['수학'], s = sizes, c = df['학년'], cmap = 'viridis', alpha = 0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.colorbar(ticks = [1,2,3], label = ['학년']) #색상에 해당
```

```python
sizes = df['학년'] * 500
plt.scatter(df['영어'], df['수학'], s = sizes, c = df['학년'], cmap = 'viridis', alpha = 0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.colorbar(ticks = [1,2,3], label = '학년', shrink = 0.5, orientation = 'horizontal') #색상에 해당하는 값과, 크기, 가로 세로결정
```
