```python
#폰트 디폴트 설정하면 좋은 것들
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'AppleGothic' #맥 설정
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시, 마이너스 글자가 꺠지는거 방지
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190,187,184]
plt.barh(labels, values) #수평으로 출력
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('x')
bar[1].set_hatch('/')
bar[2].set_hatch('.') #그래프 안 꾸미기 더 알고 싶다면 구글 ㄱㄱ
```

```python
bar = plt.bar(labels, values)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha = 'center', color = 'blue')
```
