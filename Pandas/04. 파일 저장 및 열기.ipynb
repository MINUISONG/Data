# 4. 파일 저장 및 열기

DataFrame 객체를 excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd

data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data, index = ['1번','2번','3번','4번','5번','6번','7번','8번'])
df.index.name ='지원번호'
df
```

## 저장
### 여러 파일로 저장

- CSV(Comma-Separated Values)는 데이터를 저장하고 교환하는 데 사용되는 간단한 파일 형식입니다. 각 줄이 데이터 레코드를 나타내며, 각 레코드의 필드는 쉼표(,)로 구분됩니다. 주로 텍스트 기반 파일로 저장됩니다.

```python
df.to_csv('score.csv')
```

```python
df.to_csv('socre.txt', sep='\t') # 탭으로 구분
```

```python
df.to_excel('socre.xlsx')
```

## 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows = 3) #지정된 갯수 만큼 row를 건너뜀
df
```

```python
df = pd.read_csv('score.csv', skiprows = [1,3,5]) # 1,3,5 row는 제외(row 0부터 시작)
df
```

```python
df = pd.read_csv('score.csv', nrows = 4) #처음 몇개의 Row만 가져옴
df
```

```python
df = pd.read_csv ('score.csv', skiprows = 2, nrows = 4) #처음 2개를 무시하고 그 다음을 제목컬럼으로 받아들인다음에 다음 4개의 로우만 가져옴
df
```

## 텍스트 파일 열기

```python
df = pd.read_csv('socre.txt', sep='\t') #텍스트 파일 만들때 탭으로 구분해서 탭으로 구분해서 읽어오기!
df
```

```python
df = pd.read_csv('socre.txt', sep='\t', index_col = '지원번호')
df
```

```python
df = pd.read_csv('socre.txt', sep='\t')
df.set_index('지원번호', inplace = True)
df
```
