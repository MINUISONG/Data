# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('socre.xlsx', index_col='지원번호')
df
```

## column 수정

```python
df['학교'].replace({'북산고':'상북고', '능남고' : '무슨고'}, inplace = True)
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' #학교 데이터 +등학교
df
```

## column 추가

```python
df['총합'] = df['국어'] + df ['영어'] + df['수학'] + df ['과학'] + df['사회']
df
```

```python
df['결과'] = 'F'
df
```

```python
df.loc[df['총합']>400, '결과'] = 'P'
df
```

## Coulmn 삭제

```python
df.drop(columns = ['총합'])
```

 ## Row 삭제

```python
df.drop(index = '1번')
```

```python
filt = df['수학'] <80
df.drop(index = df[filt].index)
```

## Row 추가

```python
df.loc['9번'] = ['송민의', '도래울고등학교', 185, 100,100,100,100,100, 'Mysql', 500, 'P']
```

```python
df
```

 ## cell 수정

```python
df.loc['9번',['학교', '키']] = ['동산고등학교', 200]
df
```
