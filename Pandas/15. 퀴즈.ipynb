# 15. 퀴즈

```python
import pandas as pd
data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

# 1. 전체 데이터 중에서 '영화' 정보만 출력하시오

```python
df['영화']
```

# 2. 전체 데이터 중에서 '영화', '평점' 정보만 출력하시오

```python
df[['영화', '평점']]
```

# 3. 2015년 이후에 개봉한 영화 데이터 중에서 '영화', '개봉 연도' 정보를 출력하시오.

```python
filt = df['개봉 연도'] >= 2015
df.loc[filt,['영화', '개봉 연도']]
```

df.loc는 pandas DataFrame에서 데이터를 필터링하거나 특정 행과 열을 선택하는 데 사용하는 매우 유용한 메서드입니다. 이를 이해하려면 df.loc의 구문과 동작을 단계별로 살펴보겠습니다.

기본 구문

df.loc[row_selection, column_selection]

	1.	row_selection: 행을 선택하는 조건, 슬라이스, 또는 행 이름/인덱스.
	2.	column_selection: 선택하려는 열 이름이나 열 목록.

# 4. 주어진 계산식을 참고하여 '추천 점수' column을 추가하시오

- 추천 점수 = (관객수 * 평점) // 100

```python
df['추천 점수'] = (df['관객 수'] * df['평점']) // 100
df
```
