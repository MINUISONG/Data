## 데이터 제외하기 dropna

```python
import pandas as pd
df = pd.read_excel('socre.xlsx', index_col='지원번호')
df
```

```python
df.dropna() # 전체 데이터 중에서 NaN를 포함하는 데이터 삭제 원본 데이터에 반영하고 싶다면 inplace
```

- axis : index or columns  null값이 있는 데이터를 기준으로 row를 삭제하지 column을 삭제할지 결정
- how : any or all 전체가 null이어야 지우는가 아니면 하나만 null이어도 지우는가

```python
df.dropna(axis='columns', how='any')
```

```python
df.dropna(axis='columns', how='all') #how에서 any는 디폴트값
```
